## Code and Output

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import math
```

```python
data = load_boston()
```

```python
# choosing LSTAT feature of house attributes to plot the graph
X = np.array(data['data'])[:, 12]
y = np.array(data['target'])
```

```python
plt.plot(X, y, 'b.')
plt.title('Boston Houses prices')
plt.xlabel('% Lower status of the population')
plt.ylabel('Median value of owner-occupied homes in $1000\'s')
plt.grid()
plt.show()
```

```python
plt.plot(0 * X, y, 'r.')
plt.title('Data Projected on Y-axis')
plt.xlabel('X-axis')
plt.ylabel('Median value of owner-occupied homes in $1000\'s')
plt.grid()
plt.show()
```

```python
plt.plot(X, 0 * y, 'g.')
plt.title('Data Projected on X-axis')
plt.xlabel('% Lower status of the population')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
housing_data = np.c_[X, y].T
housing_data.shape
```

```python
theta = (90 * math.pi) / 180
rotation_matrix = np.array([[np.cos(theta), np.sin(-theta)], [np.sin(theta), np.cos(theta)]])
rotation_matrix
```

```python
data_rotated = rotation_matrix.dot(housing_data)
data_rotated
```

```python
X_rotated = data_rotated[0].T
y_rotated = data_rotated[1].T
plt.plot(X_rotated, y_rotated, 'y.')
plt.title('Housing Data Rotated by 90 degrees')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
v = np.array([[0.5,0.5], [0.5, 0.5]]) #Vector Coordinates 
xy = np.dot(housing_data.T, v)
xy
```

```python
plt.plot(xy[:, 0], xy[:, 1], 'k.')
plt.title('Housing Data Projected on line x = y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```
